7.String Palindrome: Write a Python function to check if a given string is a palindrome or not.



```python
def isPalindrome(s):
  #first of all create a reversed string and if it is equal to original string it is palindrome string
  reversedStr=''
  for i in range(len(s)-1,-1,-1):
    reversedStr+=s[i]
  if reversedStr==s:
    return True
  else:
    return False
print(isPalindrome("madam"))
```

```python
#alternative solution
def isPalindrome(s):
  #first of all create a reversed string and if it is equal to original string it is palindrome string
  reversedStr=''
  for i in s:
    reversedStr=i+reversedStr
  if reversedStr==s:
    return True
  else:
    return False
print(isPalindrome("madam"))
```

```python
#alternative solution
def isPalindrome(s):
  #first of all create a reversed string and if it is equal to original string it is palindrome string
  reversedStr=''
  reversedStr=s[::-1]
  if reversedStr==s:
    return True
  else:
    return False
print(isPalindrome("madam"))
```

8. String Reversal with Slicing: Write a Python function to reverse a given string using slicing.



```python
def reversed(s):
  #first of all create a reversed string and if it is equal to original string it is palindrome string
  reversedStr=''
  reversedStr=s[::-1]
  return reversedStr
print(reversed("dhaka"))
```

9. Write a Python program that takes two strings as input and concatenates them into a single string without using the + operator.

```python
def concat(s1,s2):
  #using f string
  return f"{s1}{s2}"
print(concat("Bangladesh","Dhaka"))
```

```python
#alternative using join
def concat(s1,s2):
  return "".join([s1,s2])
print(concat("Bangladesh","Dhaka"))
```

Equilateral Triangle: All three sides are equal.
Isosceles Triangle: Any two sides are equal.
Scalene Triangle: All three sides are different.

```python
#using while loop

try:
  num=int(input("input:  "))
  x=1
  while num>0:
    x=x*num
    num=num-1
  print(x)

except ValueError:
  print("input number only")
```
