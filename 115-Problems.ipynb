3. String Reverse: Write a Python function to reverse a given string and return the reversed string.

```python
def getReverse(s):
 try:
  reversed_str=''
  for i in range(len(s)-1,-1,-1):
   reversed_str += s[i]
  return reversed_str
 except:
  return "something went wrong"

print(getReverse("Bangladesh"))
```

5.Temperature Converter: Write a Python program that converts a temperature in Celsius to Fahrenheit. Take the Celsius temperature as input from the user.  To convert Celsius to Fahrenheit we are going to use the formula “F = (C * 9/5) + 32″

```python
def celciustofarhenheit(c):
  try:
    return (c*9/5)+32
  except:
    return 0
print(f"10 celcius equals to {celciustofarhenheit(10)}F")
```

7.String Palindrome: Write a Python function to check if a given string is a palindrome or not.



```python
def isPalindrome(s):
  #first of all create a reversed string and if it is equal to original string it is palindrome string
  reversedStr=''
  for i in range(len(s)-1,-1,-1):
    reversedStr+=s[i]
  if reversedStr==s:
    return True
  else:
    return False
print(isPalindrome("madam"))
```

```python
#alternative solution
def isPalindrome(s):
  #first of all create a reversed string and if it is equal to original string it is palindrome string
  reversedStr=''
  for i in s:
    reversedStr=i+reversedStr
  if reversedStr==s:
    return True
  else:
    return False
print(isPalindrome("madam"))
```

```python
#alternative solution
def isPalindrome(s):
  #first of all create a reversed string and if it is equal to original string it is palindrome string
  reversedStr=''
  reversedStr=s[::-1]
  if reversedStr==s:
    return True
  else:
    return False
print(isPalindrome("madam"))
```

8. String Reversal with Slicing: Write a Python function to reverse a given string using slicing.



```python
def reversed(s):
  #first of all create a reversed string and if it is equal to original string it is palindrome string
  reversedStr=''
  reversedStr=s[::-1]
  return reversedStr
print(reversed("dhaka"))
```

9. Write a Python program that takes two strings as input and concatenates them into a single string without using the + operator.

```python
def concat(s1,s2):
  #using f string
  return f"{s1}{s2}"
print(concat("Bangladesh","Dhaka"))
```

```python
#alternative using join
def concat(s1,s2):
  return "".join([s1,s2])
print(concat("Bangladesh","Dhaka"))
```
